# **Pandas Parctical Questions**

## **Read Data**

```python
data = {
    "Name": ["Lina", "Omar", "Sami", "Sara", "Rami", "Tariq", "Mona", "Zain", "Leena", "Ahmed"],
    "Age": [25.0, 30.0, None, 22.0, 35.0, 40.0, 26.0, 33.0, 28.0, 30.0],
    "Gender": ["Female", "Male", "Male", "Female", "Male", None, "Female", "Male", "Female", "Male"],
    "Department": ["IT", "HR", "Marketing", "IT", "Finance", "HR", "Marketing", "IT", "Finance", "Marketing"],
    "Salary": [4500.0, 5000.0, 5200.0, 4800.0, 5600.0, 6000.0, 4700.0, 5300.0, None, 5100.0],
    "Experience": [3, 5, 2, 1, 6, 7, 4, 5, 3, 4]
}
import pandas as pd
import numpy as np

df = pd.DataFrame(data)
print(df)
```

## **Data Exploration**

```python
df = pd.DataFrame(data)
print(df)
print('-----------------------------------')
print(df.head())
print('-----------------------------------')
print(df.shape)
print('-----------------------------------')
print(df.dtypes)
print('-----------------------------------')
print(df.columns)
print('-----------------------------------')
print(df.index)
print('-----------------------------------')
print(df.isna())
print('-----------------------------------')
print(df.isna().sum())
print('-----------------------------------')
print(df.describe(include='object'))
print('-----------------------------------')
print(df.describe(include='number'))
print('-----------------------------------')
print(df.info())
```

## **Selection And Indexing**

```python
df = pd.DataFrame(data)
print(df)
print('-----------------------------------')
print(df['Name'])
print('-----------------------------------')
print(df[['Name','Salary']])
print('-----------------------------------')
print(df.iloc[3,:])
print('-----------------------------------')
print(df.loc[df["Name"] == "Leena",:])
print('-----------------------------------')
df.loc[df['Name'] == 'Omar', 'Salary'] = 4700
print(df.loc[df['Name'] == 'Omar'])
print('-----------------------------------')
print(df.query('Department == "IT"'))
print('-----------------------------------')
print(df.query('Age>26'))
print('-----------------------------------')
print(df.query('Gender == "Female"')['Salary'])
print('-----------------------------------')
print(df.query('Experience<3'))
print('-----------------------------------')
print(df.query('Department == "HR"')['Age'])
```

## **Conditional Filtering**

```python
df = pd.DataFrame(data)
print(df)
print('-----------------------------------')
print(df.query('Age>30'))
print('-----------------------------------')
print(df.query('Department == "Marketing"'))
print('-----------------------------------')
print(df.query('Department == "IT" and Gender == "Female"'))
print('-----------------------------------')
print(df.query('Salary >= 5000 and Salary <= 6000'))
print('-----------------------------------')
print(df.query('Experience >= 2 and Experience <= 5'))
print('-----------------------------------')
print(df.loc[df['Age'].isna(),:])
print('-----------------------------------')
print(df.loc[df['Experience'].isna(),:])
print('-----------------------------------')
print(df.query('Department not in ["IT"]'))
print('-----------------------------------')
print(df.query('Name.str.startswith("S")'))
print('-----------------------------------')
print(df.query('Name.str.contains("a")'))
print('-----------------------------------')
```

## **Data Modification and Creation**

```python
df = pd.DataFrame(data)
print(df)
print('-----------------------------------')
df['Bonus']=df['Salary']*0.10
print(df)
print('-----------------------------------')
df['Net Salary']=df['Salary']+df['Bonus']
print(df)
print('-----------------------------------')
def cat(age):
    if age < 30:
        return "Young"
    else:
        return "Old"

df['Age Group'] = df['Age'].apply(cat)
print(df)
print('-----------------------------------')
print(df.rename(columns={
    'Name': 'name',
    'Age': 'age',
    'Gender': 'gender',
    'Department': 'department',
    'Salary': 'salary',
    'Experience': 'experience'
}))
print('-----------------------------------')
print(df.rename(columns={
    'Salary': 'Set Salary'

}))
print('-----------------------------------')
print(df.drop('Net Salary', axis=1))
print('-----------------------------------')
print(df.tail(1))
print(df.drop(df.tail(1).index))
print('-----------------------------------')
print(df.reindex(columns=[
    'Name',
    'Age',
    'Gender',
    'Department',
    'Salary',
    'Experience'
]))

print('-----------------------------------')
df.loc[len(df)] = {'Name': 'Nora', 'Age': 29, 'Gender': 'Female', 'Department': 'IT', 'Salary': 5100, 'Experience': 4}
print(df)
print('-----------------------------------')
df['Show'] = df['Name'].str.len()
print(df)
print('-----------------------------------')
```

## **Sorting And Grouping**

```python
df = pd.DataFrame(data)
print(df)
print('-----------------------------------')
sorted_df = df.sort_values(by="Salary", ascending=True)
print(sorted_df)
print('-----------------------------------')
sorted_df = df.sort_values(by="Age", ascending=False)
print(sorted_df)
print('-----------------------------------')
result=df.groupby('Department')['Salary'].mean()
print(result)
print('-----------------------------------')
result=df.groupby('Department')['Name'].count()
print(result)
print('-----------------------------------')
result=df.groupby('Department')['Salary'].max()
print(result)
print('-----------------------------------')
result=df.groupby('Gender')['Experience'].mean()
print(result)
print('-----------------------------------')
average_salary = df.groupby(['Department', 'Gender'])['Salary'].mean()
print(average_salary)
print('-----------------------------------')
pivot_table = df.pivot_table(
    values='Salary',
    index='Department',
    columns='Gender',
    aggfunc='mean'
)

print(pivot_table)
print('-----------------------------------')
result=df.groupby(['Department', 'Gender'])['Name'].agg(['count'])
print(result)
print('-----------------------------------')
result=df.groupby('Age')['Salary'].sum()
print(result)
print('-----------------------------------')
filtered_df = df.query('Department == "IT" and Experience > 3')
count = filtered_df['Name'].count()
print(count)
print('-----------------------------------')
```

## **Handeling Missing Values**

```python
df = pd.DataFrame(data)
print(df)
print('-----------------------------------')
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df)
print('-----------------------------------')
df['Gender'] = df['Gender'].fillna('test')
print(df)
print('-----------------------------------')
df_clean = df[df.notnull().all(axis=1)]
print(df_clean)
print(df)
print('-----------------------------------')
print(df.dropna())
print('-----------------------------------')
print(df.dropna(axis=1))
print('-----------------------------------')
```

## **Advanced Operation**

```python
df = pd.DataFrame(data)
print(df)
print('-----------------------------------')
df['Sqr2']=df['Age'].apply(np.sqrt)
print(df)
print('-----------------------------------')
df['Gender']=df['Gender'].replace({'Male':0,'Female':1})
print(df)
print('-----------------------------------')
def cal(salary):
    if salary>5000:
        return salary*0.5
    else:
        return salary
df['Tax']=df['Salary'].apply(cal)
print(df)
print('-----------------------------------')
salary_range = df['Salary'].max() - df['Salary'].min()
print(salary_range)
print('-----------------------------------')
result =df['Salary'].mean()
print(result)
Q1 = df['Salary'].quantile(0.25)
print(Q1)
Q3 = df['Salary'].quantile(0.75)
print(Q3)
IQR = Q3 - Q1
print(IQR)

lower_bound = Q1 - 1.5 * IQR
print(lower_bound)
upper_bound = Q3 + 1.5 * IQR
print(upper_bound)
df = df[(df['Salary'] >= lower_bound) & (df['Salary'] <= upper_bound)]
print(df)

print('-----------------------------------')
```

## **Reshaping Data**

```python
df = pd.DataFrame(data)
print(df)
print('-----------------------------------')
result=df.pivot(index='Name',columns='Department',values='Salary')
print(result)
print('-----------------------------------')
pivot_table = df.pivot_table(
    values='Salary',
    index='Department',
    columns='Gender',
    aggfunc='mean'
)

print(pivot_table)
print('-----------------------------------')
melted_df = pd.melt(df, id_vars=['Name', 'Gender', 'Department', 'Experience'],
                    value_vars=['Age', 'Salary'],
                    var_name='Attribute',
                    value_name='Value')

print(melted_df)
print('-----------------------------------')
result=df.stack()
print(result)
print('-----------------------------------')
df_transposed = df.T
print(df_transposed)
print('-----------------------------------')
employee_count = df.groupby('Department').size().reset_index(name='EmployeeCount')

print(employee_count)
print('-----------------------------------')
df_reordered = df.reindex(columns=['Department', 'Name', 'Gender', 'Age', 'Salary'])

print(df_reordered)
print('-----------------------------------')
df_multi = df.set_index(['Department', 'Gender'])
print(df_multi)
print('-----------------------------------')
```
